```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
orders = pd.read_csv('data/olist_orders_dataset.csv')
customers = pd.read_csv('data/olist_customers_dataset.csv')
order_items = pd.read_csv('data/olist_order_items_dataset.csv')
products = pd.read_csv('data/olist_products_dataset.csv')
sellers = pd.read_csv('data/olist_sellers_dataset.csv')
payments = pd.read_csv('data/olist_order_payments_dataset.csv')
reviews = pd.read_csv('data/olist_order_reviews_dataset.csv')
geo = pd.read_csv('data/olist_geolocation_dataset.csv')
category_translation = pd.read_csv('data/product_category_name_translation.csv')


print(f"Orders: {orders.shape}")
print(f"Customers: {customers.shape}")
print(f"Order Items: {order_items.shape}")
print(f"Products: {products.shape}")
print(f"Sellers: {sellers.shape}")
print(f"Payments: {payments.shape}")
print(f"Reviews: {reviews.shape}")
print(f"Geo: {geo.shape}")
print(f"Category Translation: {category_translation.shape}")
```

```python
print("ORDERS")
print(orders.head())
print("CUSTOMERS")
print(customers.head())
print("ORDER ITEMS")
print(order_items.head())
print("PAYMENTS")
print(payments.head())
```

```python
customers['customer_city'] = customers['customer_city'].str.strip().str.lower()
customers['customer_state'] = customers['customer_state'].str.strip().str.upper()
orders['order_status'] = orders['order_status'].str.strip().str.lower()
products['product_category_name'] = products['product_category_name'].str.strip().str.lower()
sellers['seller_city'] = sellers['seller_city'].str.strip().str.lower()
sellers['seller_state'] = sellers['seller_state'].str.strip().str.upper()
geo['geolocation_city'] = geo['geolocation_city'].str.strip().str.lower()
geo['geolocation_state'] = geo['geolocation_state'].str.strip().str.upper()

print("✅ String columns normalized!")
```

```python
print("MISSING VALUES ARE:")
datasets = {
    'Orders': orders,
    'Customers': customers,
    'Order Items': order_items,
    'Products': products,
    'Payments': payments,
    'Reviews': reviews
}

for name, df in datasets.items():
    missing = df.isnull().sum().sum()
    print(f"{name}: {missing} missing values")

print("ORDERS DATA TYPES")
print(orders.dtypes)
```

```python
date_cols = ['order_purchase_timestamp', 'order_approved_at', 
             'order_delivered_carrier_date', 'order_delivered_customer_date',
             'order_estimated_delivery_date']

for col in date_cols:
    orders[col] = pd.to_datetime(orders[col])


print(orders.dtypes)
print("MISSING VALUES PER COLUMN (COLUMN FOR ORDERS")
print(orders.isnull().sum())
print("MISSING VALUES PER COLUMN (REVIEWS)")
print(reviews.isnull().sum())
print("MISSING VALUES PER COLUMN (PRODUCTS)")
print(products.isnull().sum())
```

```python

orders_delivered = orders[orders['order_status'] == 'delivered'].copy()
print(f"Total orders: {len(orders)}")
print(f"Delivered orders: {len(orders_delivered)}")

products['product_category_name'] = products['product_category_name'].fillna('unknown')

reviews['review_comment_title'] = reviews['review_comment_title'].fillna('')
reviews['review_comment_message'] = reviews['review_comment_message'].fillna('')
products = products.dropna(subset=['product_weight_g'])
products = products.merge(category_translation, on='product_category_name', how='left')
products['product_category_name_english'] = products['product_category_name_english'].fillna('unknown')
print(f"Products with English categories: {len(products)}")
```

```python
df = orders_delivered.merge(customers, on='customer_id', how='left')
df = df.merge(order_items, on='order_id', how='left')
df = df.merge(products, on='product_id', how='left')
df = df.merge(payments.groupby('order_id').agg(
    payment_type=('payment_type', 'first'),
    payment_installments=('payment_installments', 'sum'),
    payment_value=('payment_value', 'sum')
).reset_index(), on='order_id', how='left')
df = df.merge(reviews[['order_id', 'review_score', 'review_comment_message']], 
              on='order_id', how='left')

df['order_year'] = df['order_purchase_timestamp'].dt.year
df['order_month'] = df['order_purchase_timestamp'].dt.month
df['order_day'] = df['order_purchase_timestamp'].dt.day
df['order_hour'] = df['order_purchase_timestamp'].dt.hour
df['order_dayofweek'] = df['order_purchase_timestamp'].dt.day_name()

df['delivery_days'] = (df['order_delivered_customer_date'] - 
                       df['order_purchase_timestamp']).dt.days
print(f"Shape: {df.shape}")
print(f"\nColumns: {list(df.columns)}")
```

```python
# Normalize string columns
customers['customer_city'] = customers['customer_city'].str.strip().str.lower()
customers['customer_state'] = customers['customer_state'].str.strip().str.upper()
orders['order_status'] = orders['order_status'].str.strip().str.lower()
sellers['seller_city'] = sellers['seller_city'].str.strip().str.lower()
sellers['seller_state'] = sellers['seller_state'].str.strip().str.upper()

print("✅ String columns normalized!")
```

```python
# Add time of day category
conditions = [
    df['order_hour'].between(4, 11),
    df['order_hour'].between(12, 16),
    df['order_hour'].between(17, 22),
    df['order_hour'].isin([23, 0, 1, 2, 3])
]
choices = ['Morning', 'Afternoon', 'Evening', 'Night']
df['time_of_day'] = np.select(conditions, choices, default='Other')

# Add delay flag
df['is_delayed'] = np.where(
    df['order_delivered_customer_date'] > df['order_estimated_delivery_date'],
    'Delayed', 'Not Delayed'
)

print(df['time_of_day'].value_counts())
print("\n", df['is_delayed'].value_counts())
```

```python
# Add delivery time breakdown
df['from_approved_to_carrier_days'] = (
    df['order_delivered_carrier_date'] - df['order_approved_at']
).dt.total_seconds() / (24 * 3600)

df['from_carrier_to_customer_days'] = (
    df['order_delivered_customer_date'] - df['order_delivered_carrier_date']
).dt.total_seconds() / (24 * 3600)

df['delivery_delay_days'] = (
    df['order_delivered_customer_date'] - df['order_estimated_delivery_date']
).dt.total_seconds() / (24 * 3600)

# Weekend vs Weekday
df['day_type'] = df['order_dayofweek'].apply(
    lambda x: 'Weekend' if x in ['Saturday', 'Sunday'] else 'Weekday'
)

# Season (Southern Hemisphere)
def get_season(month):
    if month in [12, 1, 2]: return 'Summer'
    elif month in [3, 4, 5]: return 'Autumn'
    elif month in [6, 7, 8]: return 'Winter'
    else: return 'Spring'

df['season'] = df['order_month'].apply(get_season)

print(df['day_type'].value_counts())
print("\n", df['season'].value_counts())
print("\n✅ New features added!")
```

```python
print(f"Total Orders: {df['order_id'].nunique():,}")
print(f"Total Customers: {df['customer_unique_id'].nunique():,}")
print(f"Total Products: {df['product_id'].nunique():,}")
print(f"Total Sellers: {df['seller_id'].nunique():,}")
print(f"Total Revenue: R$ {df['payment_value'].sum():,.2f}")
print(f"Average Order Value: R$ {df.groupby('order_id')['payment_value'].sum().mean():,.2f}")
print(f"Average Delivery Days: {df['delivery_days'].mean():.1f} days")
print(f"Average Review Score: {df['review_score'].mean():.2f} / 5")
print(f"\nDate Range: {df['order_purchase_timestamp'].min().date()} to {df['order_purchase_timestamp'].max().date()}")
```

```python
from datetime import datetime

reference_date = df['order_purchase_timestamp'].max() + pd.Timedelta(days=1)

rfm = df.groupby('customer_unique_id').agg(
    recency=('order_purchase_timestamp', lambda x: (reference_date - x.max()).days),
    frequency=('order_id', 'nunique'),
    monetary=('payment_value', 'sum')
).reset_index()

rfm['r_score'] = pd.qcut(rfm['recency'], q=4, labels=[4, 3, 2, 1])
rfm['f_score'] = pd.qcut(rfm['frequency'].rank(method='first'), q=4, labels=[1, 2, 3, 4])
rfm['m_score'] = pd.qcut(rfm['monetary'], q=4, labels=[1, 2, 3, 4])

rfm['rfm_score'] = rfm['r_score'].astype(str) + rfm['f_score'].astype(str) + rfm['m_score'].astype(str)
def segment(row):
    r = int(row['r_score'])
    f = int(row['f_score'])
    m = int(row['m_score'])
    if r >= 3 and f >= 3:
        return 'Champions'
    elif r >= 3 and f >= 2:
        return 'Loyal Customers'
    elif r >= 3:
        return 'Recent Customers'
    elif r >= 2 and f >= 2:
        return 'Potential Loyalists'
    elif r == 1 and f >= 3:
        return 'At Risk'
    elif r == 1 and f == 1:
        return 'Lost'
    else:
        return 'Hibernating'

rfm['segment'] = rfm.apply(segment, axis=1)
segment_counts = rfm['segment'].value_counts().reset_index()
segment_counts.columns = ['segment', 'count']

fig, ax = plt.subplots(figsize=(12, 6))
colors = ['#2ecc71', '#27ae60', '#3498db', '#f39c12', '#e74c3c', '#c0392b', '#95a5a6']
bars = ax.bar(segment_counts['segment'], segment_counts['count'],
              color=colors, edgecolor='white')

for bar, val in zip(bars, segment_counts['count']):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 100,
            f'{val:,}\n({val/len(rfm)*100:.1f}%)',
            ha='center', fontsize=9)

ax.set_title('Customer Segments — RFM Analysis', fontsize=14, fontweight='bold')
ax.set_xlabel('Segment')
ax.set_ylabel('Number of Customers')
plt.xticks(rotation=15)
plt.tight_layout()
plt.savefig('rfm_segments.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n RFM Segment Summary:")
print(rfm.groupby('segment').agg(
    customers=('customer_unique_id', 'count'),
    avg_monetary=('monetary', 'mean'),
    avg_recency=('recency', 'mean')
).round(2).sort_values('customers', ascending=False))
```

```python
# NPS Analysis
nps_data = df.groupby('review_score')['order_id'].nunique().reset_index()
nps_data.columns = ['score', 'count']

total = nps_data['count'].sum()
promoters = nps_data[nps_data['score'] == 5]['count'].sum()
detractors = nps_data[nps_data['score'].isin([1, 2, 3])]['count'].sum()
nps = ((promoters - detractors) / total) * 100

print(f"Promoters (5 stars): {promoters:,} ({promoters/total*100:.1f}%)")
print(f"Passives (4 stars): {nps_data[nps_data['score']==4]['count'].sum():,}")
print(f"Detractors (1-3 stars): {detractors:,} ({detractors/total*100:.1f}%)")
print(f"\n⭐ NPS Score: {nps:.1f}")

if nps >= 50:
    print("Excellent! 🟢")
elif nps >= 0:
    print("Good 🟡")
else:
    print("Needs improvement 🔴")
```

```python
# Delivery stages analysis
df['approval_days'] = (df['order_approved_at'] - 
                       df['order_purchase_timestamp']).dt.total_seconds() / (24*3600)
df['carrier_days'] = (df['order_delivered_carrier_date'] - 
                      df['order_approved_at']).dt.total_seconds() / (24*3600)
df['customer_days'] = (df['order_delivered_customer_date'] - 
                       df['order_delivered_carrier_date']).dt.total_seconds() / (24*3600)

avg_approval = df['approval_days'].mean()
avg_carrier = df['carrier_days'].mean()
avg_customer = df['customer_days'].mean()
total = avg_approval + avg_carrier + avg_customer

fig, ax = plt.subplots(figsize=(10, 6))
stages = ['Purchase to\nApproval', 'Approval to\nCarrier', 'Carrier to\nCustomer']
values = [avg_approval, avg_carrier, avg_customer]
colors = ['#3498db', '#e67e22', '#2ecc71']
bars = ax.bar(stages, values, color=colors, edgecolor='white', width=0.5)

for bar, val in zip(bars, values):
    pct = val/total*100
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1,
            f'{val:.1f} days\n({pct:.0f}%)', ha='center', fontsize=10, fontweight='bold')

ax.set_title('Average Days at Each Delivery Stage', fontsize=14, fontweight='bold')
ax.set_ylabel('Average Days')
plt.tight_layout()
plt.savefig('delivery_stages.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"Purchase → Approval: {avg_approval:.1f} days ({avg_approval/total*100:.0f}%)")
print(f"Approval → Carrier: {avg_carrier:.1f} days ({avg_carrier/total*100:.0f}%)")
print(f"Carrier → Customer: {avg_customer:.1f} days ({avg_customer/total*100:.0f}%)")
print("✅ Plot saved!")
```
